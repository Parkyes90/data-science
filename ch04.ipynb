```python
def heaviside_step(x):
    if isinstance(x, np.ndarray):
        return np.where(x >= 0, 1, 0)
    return 1.0 if x >= 0 else 0.0
heaviside_step(-0.0001)
```

```python
heaviside_step(0)
```

```python
heaviside_step(0.0001)
```

```python
def f(x):
    return x ** 3 - 3* x ** 2 + x
```

```python
x = np.linspace(-1, 3, 400)
y = f(x)
```

```python
y
```

```python
plt.plot(x,y, 'ro-')
plt.xlim(-2, 4)
plt.title('함수 $f(x) x= x^3 - 3x^2 + x$의 그래프')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5,4))
plt.show()
```

```python
xx = np.linspace(-2,2, 100)
yy = np.exp(xx)
plt.plot(xx, yy)
plt.axhline(1, c='r', ls='--')
plt.axhline(0, c='r', ls='--')
plt.axvline(0, c='r', ls='--')
plt.xlabel('$x$')
plt.ylabel('$\exp(x)$')
plt.show()
```

```python
def logistic(x):
    return 1/ (1+np.exp(-x))

xx = np.linspace(-10, 10, 100)
plt.plot(xx, logistic(xx))
plt.show()
```

```python
def f(x,y):
    return 2 * x **2 + 6*x *y + 7 *y**2 - 26 *x - 54*y + 107
xx = np.linspace(-3,7,100)
yy = np.linspace(-3,7,100)
```

```python
X, Y = np.meshgrid(xx, yy)
Z = f(X,Y)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z, linewidth=0.1)
ax.view_init(40, -110)
plt.show()
```

```python
def softmax(x,w):
    e = np.exp(w * x)
    return np.exp(w *x) / e.sum()
x = [2.0, 1.0, 0.5]
y = softmax(x, np.ones(3))
```
